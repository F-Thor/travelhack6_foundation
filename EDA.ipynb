## Исследуем людей, которые ездили несколько раз

```python
df.columns
```

```python
unique_clients_per_req = df.drop_duplicates().groupby('Заявка № ').apply(lambda dff: dff['ИД клиента'].nunique() == len(dff))
(~unique_clients_per_req).sum()
```

```python
unique_clients_per_req[~unique_clients_per_req].index
```

```python
df.shape
```

```python
df.drop_duplicates().shape
```

```python
df[df['Заявка № '].isin(unique_clients_per_req[~unique_clients_per_req].index)]\
.sort_values(['Заявка № ', 'ИД клиента'])
```

```python
t = df.groupby('ИД клиента').size()
t.name = 'колво поездок'
print(t.value_counts())
multi_client = t[t>1].index
```

```python
df = df.merge(t.reset_index(), how='left')
```

```python
t = df.groupby('Заявка № ').size()
t.name = 'Туристов_расчет'
df = df.merge(t.reset_index(), how='left')
```

```python
plt.scatter(df['Год'], df['Глубина продаж']);
```

```python
df.loc[df['Туристов'] != df['Туристов_расчет']]['Заявка № '].nunique()
```

```python
df.loc[df['Туристов'] != df['Туристов_расчет'], ['ИД клиента', 'Заявка № ', 'Туристов', 'Туристов_расчет']]\
.sort_values(['Заявка № ', 'ИД клиента'])\
.tail(20)
```

```python
for p in sorted(df['колво поездок'].unique()):
    print(p, 'поездки')
    print(df[df['колво поездок']==p].groupby(['ИД клиента'])['Страна тура'].nunique().value_counts().sort_index(), end='\n\n')
```
